```python
# importing all the necessary modules:-
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import statsmodels.api as sm

from sklearn.metrics import accuracy_score, confusion_matrix
```

### Exploring the dataset:

```python
# Reading the dataset:
df = pd.read_csv("diabetes.csv")
df
```

```python
df.info()
```

```python
df.shape
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# adding a constant to the independent variables:
# we are adding a constant because as the line equation is y=mx+c so c is a constant which is not present in the statsmodel api.
x = sm.add_constant(x)
```

```python
x
```

## Splitting the dataset into Train and Test:

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=1)
print(f"Train Data: {xtrain.shape}, {ytrain.shape}")
print(f"Test Data: {xtest.shape}, {ytest.shape}")
```

## Fitting the Logistic Regression Model:-

```python
logistic_model = sm.Logit(ytrain,xtrain) #initializing the model
trained_logistic_model = logistic_model.fit() #train the model
print(trained_logistic_model.summary())
```

```python
predicted_probabilities = trained_logistic_model.predict(xtest)
print(predicted_probabilities)
```

```python
threshold = 0.5
predictions = (predicted_probabilities > threshold).astype(int)
print(predictions)
```

```python
print(f"Accuracy of the model: {accuracy_score(ytest, predictions)}")
print(confusion_matrix(ytest,predictions))
```
